```python
# import necessary libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
```

```python
# load the merged train dataset
train = pd.read_pickle("../data/train_merged.pkl")

print(train.shape)
```

```python
# missing values analysis
missing_percent = (
    train.isnull().sum()
    / len(train)
) * 100

missing_percent = missing_percent.sort_values(
    ascending=False
)

missing_percent.head(20)
```

```python
# Droo extremly sparse columns with more than 95% missing values
drop_cols = missing_percent[
    missing_percent > 95
].index

print("Columns to Drop:", len(drop_cols))
```

```python
list(drop_cols)
```

```python
# dropping the above columns
train = train.drop(
    columns=drop_cols
)

print(train.shape)
```

```python
# Seperate numerical and categorical
categorical_cols = (
    train.select_dtypes(
        include=["object"]
    ).columns
)

numerical_cols = (
    train.select_dtypes(
        exclude=["object"]
    ).columns
)

print("Categorical:", len(categorical_cols))
print("Numerical:", len(numerical_cols))
```

```python
# Missing value in categorical columns
train[categorical_cols].isnull().sum().sort_values(
    ascending=False
).head(20)
```

```python
# Missing values in numerical columns
train[numerical_cols].isnull().sum().sort_values(
    ascending=False
).head(20)
```

```python
# Strategy for missing values
num_missing = (
    train[numerical_cols]
    .isnull()
    .sum()
    .sort_values(ascending=False)
)

num_missing.head(10)
```

```python
# Categorical missing values
cat_missing = (
    train[categorical_cols]
    .isnull()
    .sum()
    .sort_values(ascending=False)
)

cat_missing.head(10)
```

```python
# we are going to fill missing values in numerical columns with median.
train["DeviceType_missing"] = train["DeviceType"].isnull().astype(int)
```

```python
# Filling the numerical columns
for col in numerical_cols:
    train[col] = train[col].fillna(
        train[col].median()
    )
```

```python
# Filling Categorical columns
for col in categorical_cols:
    train[col] = train[col].fillna(
        "Missing"
    )
```

```python
# Verifying the missingvalues
train.isnull().sum().sum()
```

```python
# Saving the cleaned dataset
train.to_pickle(
    "../data/train_clean.pkl"
)

print("Saved Successfully")
```
