```python
import pandas as pd
import numpy as np

train_transaction = pd.read_csv("../data/train_transaction.csv")
train_identity = pd.read_csv("../data/train_identity.csv")

print("Train Transaction Shape:", train_transaction.shape)
print("Train Identity Shape:", train_identity.shape)
```

```python
# Understanding the data
train_transaction.head()
```

```python
train_transaction.info()
```

```python
# Understanding the target variable
train_transaction["isFraud"].value_counts()
```

```python
train_transaction["isFraud"].value_counts(normalize=True)*100
```

We can observe the above cell, that fraudulent and non-fraudulent transactions ratio is 96.51 and 3.49 , which usual in the current banking sector and most transactions are genuine.

```python
# Visualizing fraud dsitribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

sns.countplot(
    x="isFraud",
    data=train_transaction
)

plt.title("Fraudulent vs Non-Fraudulent Transactions")
plt.show()
```

```python
# Transaction Amount analysis
train_transaction["TransactionAmt"].describe()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    train_transaction["TransactionAmt"],
    bins=100
)

plt.title("Transaction Amount Distribution")
plt.show()
```

Above we can see right skewed distribution

```python
# Fraud by transaction amount
plt.figure(figsize=(10,5))

sns.boxplot(
    x="isFraud",
    y="TransactionAmt",
    data=train_transaction
)

plt.show()
```

```python
# Missing values analysis
missing_values = train_transaction.isnull().sum()

missing_values = missing_values[
    missing_values > 0
]

missing_values = (
    missing_values
    .sort_values(ascending=False)
)

missing_values.head(10)
```
