```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_pickle("../data/train_clean.pkl")

print(train.shape)
```

```python
#1. Log transaction amount
train["TransactionAmt_log"] = np.log1p(
    train["TransactionAmt"]
)
```

```python
#2. email domain feture
train["P_email_prefix"] = (
    train["P_emaildomain"]
    .astype(str)
    .str.split(".")
    .str[0]
)
```

```python
#3. Mobile device flag
train["is_mobile"] = (
    train["DeviceType"] == "mobile"
).astype(int)
```

```python
# Transaction Amount bins
train["Amt_Bin"] = pd.qcut(
    train["TransactionAmt"],
    q=10,
    duplicates="drop"
)
```

```python
# Card frequency feature
train["card1_frequency"] = (
    train.groupby("card1")["card1"]
    .transform("count")
)
```

```python
# Frequency Feature
train["card1_frequency"] = (
    train.groupby("card1")["card1"]
    .transform("count")
)

train["email_frequency"] = (
    train.groupby("P_emaildomain")["P_emaildomain"]
    .transform("count")
)
```

```python
train[
    [
        "TransactionAmt",
        "TransactionAmt_log",
        "P_emaildomain",
        "P_email_prefix",
        "DeviceType",
        "is_mobile",
        "Amt_Bin",
        "card1_frequency",
        "email_frequency"
    ]
].head()
```

```python
# Email domain frequency feature
train["email_frequency"] = (
    train.groupby("P_emaildomain")
    ["P_emaildomain"]
    .transform("count")
)
```

```python
train[
[
"TransactionAmt",
"TransactionAmt_log",
"is_mobile",
"card1_frequency",
"email_frequency"
]
].head()
```

```python
# saving the feature engineered dataset
train.to_pickle(
    "../data/train_feature_engineered.pkl"
)

print("Feature Engineering Saved")
```
